```python
with open(r'C:\Users\Jeonghoon Kong\OneDrive\Desktop\Python\Python_coursera\Course 1 - Notebook Resources\resources\week-1\datasets\ferpa.txt') as file:
    wiki = file.read()

wiki
```

```python
re.findall('[\w]*\[edit\]', wiki)


## \ : escape speical Character
## [\w] : any characters, digits
## * : 0 or more times
## [] : 요거는 보기 더러우니깐 해주는거다!!!
```

```python
for title in re.findall('[\w ]*\[edit\]', wiki):
    print(
        re.split('[\[]', title)[0]
    )
```

```python
re.findall('([\w ]*)(\[edit\])', wiki)
```

```python
for item in re.finditer("([\w ]*)(\[edit\])", wiki):
    print(item.groups())
```

```python
for item in re.finditer('(?P<title>[\w ]*)(?P<edit_link>\[edit\])', wiki):
    print(item.groupdict()['title'])
```

```python
print(item.groupdict())
```
