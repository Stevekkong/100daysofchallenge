```python
cylinders = set(d['cyl'] for d in mpg)
```

```python
cty_mpg_by_cyl = [] #Empty List

for c in cylinders: #iterate each type of cylinder
    sum_mpg = 0
    cyl_type_count = 0
    for d in mpg: #iterate all dictionaries in mpg
        if d['cyl'] == c:
            sum_mpg += float(d['cty'])
            cyl_type_count +=1
    
    cty_mpg_by_cyl.append((c, sum_mpg/cyl_type_count)) #append the list


cty_mpg_by_cyl.sort(key=lambda x:x[0])
print(cty_mpg_by_cyl)
```

```python
#unique values for class type

print(mpg[0].keys()
)

vehicle_class = set(d['class'] for d in mpg)
print(vehicle_class)
```

```python
# And here's an example of how to find the average hwy mpg for each class of vehicle in our dataset.
```

```python

sum_classes=[]

for c in vehicle_class:
    sum_class=0
    class_count = 0
    for d in mpg:
        if d['class'] ==c:
            sum_class += float(d['hwy'])
            class_count += 1
        
    sum_classes.append((c, sum_class/class_count))
            
sum_classes
```

```python
sum_classes.sort(key=lambda x:x[0])
sum_classes
```
