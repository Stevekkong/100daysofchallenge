# Objects and map()

```python
class Person:
    department = 'Chemical Engineering'
    
    def set_name(self, new_name):
        self.name = new_name
    def set_location(self, new_location):
        self.location = new_location
```

```python
person = Person()
person.set_name('Steve Kong')
person.set_location('London, ON, Canada')
print('{} lives in {} and works in the department of {}'.format(person.name, person.location, person.department))
```

```python
person.name
person.location
```

```python
store1 = [10.00, 11.00, 12.34, 2.34]
store2 = [9.00, 11.10, 12.34, 2.01]

cheapest = map(min, store1, store2)
cheapest
```

```python
for item in cheapest:
    print(item)
```

# Lambda and List Comprehension

```python
my_function = (lambda a, b, c : a+b)
```

```python
my_function(1,2,3)
```

```python
my_list = []
for number in range(0, 1000):
    if number %2 ==0:
        my_list.append(number)
        
my_list
```

```python
uneven_list = []

for number in range(0,50):
    if number %2 != 0:
        uneven_list.append(number)

print(uneven_list)
```

```python
my_list = [number for number in range(0,50) if number%2 ==0]
```

```python
my_new_list = [num for num in range(0,1000) if num %2 !=0]
my_new_list
```

# Regex Pattern

```python
grades="BCACAAAABCBCBAA CAA"

re.findall('B', grades)
```

```python
re.findall('[AB]', grades)
```

```python
re.findall('[A][B-C]', grades)
```

```python
re.findall('AB|AC|BC', grades)
```

```python
re.findall('[^B]', grades)
```

```python
re.findall('^[^A]', grades)
```

```python
re.findall('A{2,5}', grades)
```

```python
re.findall('C{1,10}B{1,10}A{1,10}', grades)
```
